#O(logn)

```python
import time
import matplotlib.pyplot as plt

def binary_search(arr, target):
    left = 0
    right = len(arr) - 1

    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Function to measure the execution time of binary search
def measure_time(arr, target):
    start_time = time.time()
    binary_search(arr, target)
    end_time = time.time()
    return end_time - start_time

# Main code to input different numbers, run binary search, and plot the runtimes
input_sizes = [10**i for i in range(1, 7)]  # Vary the input sizes
runtimes = []

for size in input_sizes:
    arr = list(range(size))
    target = size // 2  # Target is chosen as mid-point for simplicity
    runtime = measure_time(arr, target)
    runtimes.append(runtime)

# Plotting the runtimes
plt.plot(input_sizes, runtimes, marker='o')
plt.xlabel('Input Size (n)')
plt.ylabel('Runtime (seconds)')
plt.title('Runtime of Binary Search')
plt.xscale('log')  # Logarithmic scale for x-axis
plt.yscale('log')  # Logarithmic scale for y-axis
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt

def largest_power_of_2(num):
    power = 0
    while num >= 2:
        num //= 2
        power += 1
    return 2 ** power

# Function to measure the execution time of finding the largest power of 2
def measure_time(num):
    start_time = time.time()
    largest_power_of_2(num)
    end_time = time.time()
    return (end_time - start_time) * 1000  # Convert to milliseconds

# Main code to input different numbers, find the largest power of 2, and plot the runtimes
input_sizes = [10**i for i in range(1, 7)]  # Vary the input sizes
runtimes = []

for size in input_sizes:
    runtime = measure_time(size)
    runtimes.append(runtime)

# Plotting the runtimes
plt.plot(input_sizes, runtimes, marker='o')
plt.xlabel('Input Size (n)')
plt.ylabel('Runtime (milliseconds)')
plt.title('Runtime of Finding Largest Power of 2')
plt.grid(True)
plt.show()
```

```python
import time
import matplotlib.pyplot as plt


def find_max(arr):
    max_element = arr[0]
    for num in arr[1:]:
        if num > max_element:
            max_element = num
    return max_element

# Function to measure the execution time of finding the maximum element
def measure_time(arr):
    start_time = time.time()
    find_max(arr)
    end_time = time.time()
    return (end_time - start_time) * 1000  # Convert to milliseconds

# Main code to input different array sizes, find the maximum element, and plot the runtimes
input_sizes = [1,2,3,6,14,20,40,50,64,100]  # Vary the input sizes
runtimes = []

for size in input_sizes:
    arr = [1,2,3,4,5,6,8,9,7,14,15,20,40,45,50,53,64,66,77,100]
    runtime = measure_time(arr)
    runtimes.append(runtime)

# Plotting the runtimes
plt.plot(input_sizes, runtimes, marker='o')
plt.xlabel('Input Size (n)')
plt.ylabel('Runtime (milliseconds)')
plt.title('Runtime of Finding Maximum Element (O(n))')
plt.grid(True)
plt.show()
```

# O(n)

```python
import time
import matplotlib.pyplot as plt

def sum_of_range_num(n):
    sum = 0
    for i in range(0, n):
        sum = sum + i
    return sum


take_input = [10, 50, 150, 500,650,700]
exe_time = []
for i in take_input:
    start = time.time()
    result = sum_of_range_num(i)
    end = time.time()
    exe_time.append((end-start)*1000)
    print(f"Input size: {i}, Runtime: {(end-start)*1000:.4f} milliseconds")
# Plot the graph
plt.plot(take_input, exe_time)
plt.title('Graph')
plt.xlabel('Input Size')
plt.ylabel('Execution Time(milliseconds)')
plt.grid(True)
plt.show()
```

# O(nlogn)

```python
import random
import time
import matplotlib.pyplot as plt

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            temp = arr[i]
            arr[i] = arr[j]
            arr[j] = temp
    temp = arr[i + 1]
    arr[i + 1] = arr[high]
    arr[high] = temp
    return i + 1

def quicksort(arr, low, high):
    if low < high:
        pivotIndex = partition(arr, low, high)
        quicksort(arr, low, pivotIndex - 1)
        quicksort(arr, pivotIndex + 1, high)


def measure_execution_time(take_input):
    execution_times = []
    for i in take_input:
        #generate random array
        arr = [random.randint(1,1000) for _ in range(i)]
        start = time.time()
        quicksort(arr,0,len(arr)-1)
        end=time.time()
        execution_time = (end-start) * 1000
        execution_times.append(execution_time)
        print(f"Input {i} Execution time: {execution_time:.4f} ms")
    return execution_times


take_input = [500,1000,1500,2000,2500,3000,3500,4000]
execution_times = measure_execution_time(take_input)

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.plot(take_input, execution_times)
plt.title('O(nlogn)')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (milliseconds)')
plt.xscale('log')
plt.grid(True)
plt.show()
```

#O(n^2)

```python
import matplotlib.pyplot as plt
import time

def quadratic_time(n):
    start = time.time()
    result = 0
    for i in range(n):
        for j in range(n):
            result += i * j
    end = time.time()
    return (end - start) * 1000

arr = [10, 50, 150, 500,650,700,850 ,1000]
execution_times = []
for i in arr:
    execution_time = quadratic_time(i)
    execution_times.append(execution_time)
    print(f"Input size: {i}, Runtime: {execution_time:.4f} milliseconds")

plt.plot(arr, execution_times)
plt.title('O(n^2)')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (milliseconds)')
plt.xscale('log')
plt.show()
```
