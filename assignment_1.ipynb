# O(n)

```python
import time
import matplotlib.pyplot as plt

def sum_of_range_num(n):
    sum = 0
    for i in range(0, n):
        sum = sum + i
    return sum


take_input = [10, 50, 150, 500,650,700]
exe_time = []
for i in take_input:
    start = time.time()
    result = sum_of_range_num(i)
    end = time.time()
    exe_time.append((end-start)*1000)
    print(f"Input size: {i}, Runtime: {(end-start)*1000:.4f} milliseconds")
# Plot the graph
plt.plot(take_input, exe_time)
plt.title('Graph')
plt.xlabel('Input Size')
plt.ylabel('Execution Time(milliseconds)')
plt.grid(True)
plt.show()
```

#O(logn)

```python
import matplotlib.pyplot as plt
import time

def logarithmic_time(n):
    start = time.time()
    i = 1
    while i < n:
        i *= 2
    end = time.time()
    return (end - start) * 1000

arr = [100, 10000, 50000, 100000, 150000, 200000]
execution_times = []
for i in arr:
    execution_time = logarithmic_time(i)
    execution_times.append(execution_time)
    print(f"Input size: {i}, Runtime: {execution_time:.4f} milliseconds")

plt.plot(arr, execution_times)
plt.title('O(log n)')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (milliseconds)')
plt.show()
```

# O(nlogn)

```python
import matplotlib.pyplot as plt
import time

def n_logarithmic_time(n):
    start = time.time()
    result = 0
    for i in range(n):
        for j in range(1, n, 2):
            result += i * j
    end = time.time()
    return (end - start) * 1000

arr = [10, 100, 250, 500, 650, 700, 850, 1000]
execution_times = []
for i in arr:
    execution_time = n_logarithmic_time(i)
    execution_times.append(execution_time)
    print(f"Input size: {i}, Runtime: {execution_time:.4f} milliseconds")

plt.plot(arr, execution_times)
plt.title('O(n log n)')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (milliseconds)')
plt.show()
```

#O(n^2)

```python
import matplotlib.pyplot as plt
import time

def quadratic_time(n):
    start = time.time()
    result = 0
    for i in range(n):
        for j in range(n):
            result += i * j
    end = time.time()
    return (end - start) * 1000

arr = [10, 50, 150, 500,650,700,850 ,1000]
execution_times = []
for i in arr:
    execution_time = quadratic_time(i)
    execution_times.append(execution_time)
    print(f"Input size: {i}, Runtime: {execution_time:.4f} milliseconds")

plt.plot(arr, execution_times)
plt.title('O(n^2)')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (milliseconds)')
plt.xscale('log')
plt.show()
```
